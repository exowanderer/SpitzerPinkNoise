```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

# DFM GP Model named Celerite

# SKlearn

```python
from sklearn.gaussian_process import GaussianProcess
```

```python
np.random.seed(42)
```

```python
def f(x):
    """The function to predict."""
    return x * np.sin(x)
```

```python
X = np.linspace(0.1, 9.9, 20)
X = np.atleast_2d(X).T

# Observations and noise
y = f(X).ravel()
dy = 0.5 + 1.0 * np.random.random(y.shape)
noise = np.random.normal(0, dy)
y += noise

# Mesh the input space for evaluations of the real function, the prediction and
# its MSE
x = np.atleast_2d(np.linspace(0, 10, 1000)).T

# Instanciate a Gaussian Process model
gp = GaussianProcess(corr='squared_exponential', theta0=1e-1,
                     thetaL=1e-3, thetaU=1,
                     nugget=(dy / y) ** 2,
                     random_start=100)

# Fit to data using Maximum Likelihood Estimation of the parameters
gp.fit(X, y)

# Make the prediction on the meshed x-axis (ask for MSE as well)
y_pred, MSE = gp.predict(x, eval_MSE=True)
sigma = np.sqrt(MSE)

# Plot the function, the prediction and the 95% confidence interval based on
# the MSE
fig = plt.figure(figsize=(16,12))
plt.plot(x, f(x), 'r:', label=u'$f(x) = x\,\sin(x)$')
plt.errorbar(X.ravel(), y, dy, fmt='r.', markersize=10, label=u'Observations')
plt.plot(x, y_pred, 'b-', label=u'Prediction')
plt.fill(np.concatenate([x, x[::-1]]),
        np.concatenate([y_pred - 1.9600 * sigma,
                       (y_pred + 1.9600 * sigma)[::-1]]),
        alpha=.5, fc='b', ec='None', label='95% confidence interval')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.ylim(-10, 20)
plt.legend(loc='upper left')
```

```python
import celerite
from celerite import terms
ExpSquaredKernel = terms.RealTerm
ExpSine2Kernel   = terms.SHOTerm

# Set up the Gaussian process.
kernel1 = ExpSquaredKernel(1.0, 1.0)
kernel2 = ExpSine2Kernel(1.0, 4*np.pi)

gp1 = celerite.GP(kernel1)
gp2 = celerite.GP(kernel2)
gp3 = celerite.GP(kernel1 + kernel2)

# Pre-compute the factorization of the matrix.
gp1.compute(X.ravel(), noise)
gp2.compute(X.ravel(), noise)
gp3.compute(X.ravel(), noise)

# Compute the log likelihood.
print(gp1.lnlikelihood(y))
print(gp2.lnlikelihood(y))
print(gp3.lnlikelihood(y))
```

```python
t = np.linspace(0, 10, 500)

mu1, cov1 = gp1.predict(y, t)
mu2, cov2 = gp2.predict(y, t)
mu3, cov3 = gp3.predict(y, t)
std1 = np.sqrt(np.diag(cov1))
std2 = np.sqrt(np.diag(cov2))
std3 = np.sqrt(np.diag(cov3))
```

```python
plt.figure(figsize=(16,12))
plt.plot(x, f(x), 'r:', label=u'$f(x) = x\,\sin(x)$')
plt.errorbar(X.ravel(), y, dy, fmt='r.', markersize=10, label=u'Observations')

plt.fill(np.concatenate([t,t[::-1]]), np.concatenate([mu1-2*std1, (mu1+2*std1)[::-1]])\
                                                     , color='grey'  , label='celerite Prediction 1')
plt.fill(np.concatenate([t,t[::-1]]), np.concatenate([mu2-2*std2, (mu2+2*std2)[::-1]])\
                                                     , color='orange', label='celerite Prediction 2')
plt.fill(np.concatenate([t,t[::-1]]), np.concatenate([mu3-2*std3, (mu3+2*std3)[::-1]])\
                                                     , color='pink'  , label='celerite Prediction 3')
plt.fill(np.concatenate([x, x[::-1]]),
         np.concatenate([y_pred - 1.9600 * sigma,
                       (y_pred + 1.9600 * sigma)[::-1]]),
        alpha=.5, fc='b', ec='None', label='95% confidence interval')

plt.plot(t, mu1   , color='grey', label=u'celerite Prediction 1')
plt.plot(t, mu2   , color='grey', label=u'celerite Prediction 2')
plt.plot(t, mu3   , color='grey', label=u'celerite Prediction 3')
plt.plot(x, y_pred, 'b-', label=u'SKLearn Prediction')

plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.ylim(-10, 20)
plt.legend(loc='upper left')
```
